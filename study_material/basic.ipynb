```python
## 만들고자 하는 함수 문자열 데이값이 주지면 년월일을 제시해주는 연습

class MyClass:
    name = "MyName"
    def print(self, str):
        print(self.name + '' + str)

x = MyClass()
x.print("Hello, Python")
```

```python
''' lambda 람다

lambda 매개변수 : 표현식'''

(lambda x,y: x+y)(10,20)
```

```python
print((lambda a : str(a)[0:4])(20220612)) ### 데이터에서 날짜만 가져옴
(lambda b :  str(b)[4:6])(20220612) ### 데이터에서 월만 가져옴
```
